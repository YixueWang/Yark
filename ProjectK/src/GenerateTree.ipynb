```python
news_sequence = pd.read_csv('news_sequence_05.csv', sep='	', names=['ID', 'Sequence', 'Confidence'], header=True)
```

```python
news_sequence['Sequence'] = news_sequence['Sequence'].apply(lambda x : x.replace("{",""))
news_sequence['Sequence'] = news_sequence['Sequence'].apply(lambda x : x.replace("}",""))
news_sequence['Sequence'] = news_sequence['Sequence'].apply(lambda x : x.replace("<",""))
news_sequence['Sequence'] = news_sequence['Sequence'].apply(lambda x : x.replace(">",""))
```

```python
news_id_list = news_sequence['ID'].tolist()
news_seq_list = news_sequence['Sequence'].tolist()
news_seq_nested = []
```

```python
for seq in news_seq_list :
    seq = seq.split(",")
    news_seq_nested.append(seq)
```

```python
keyword_map = {}

with open('keyword_list.csv', 'rb') as f:
    reader = csv.reader(f)
    reader.next() #skipheader
    
    for row in reader:
        keyword_id = row[0]
        keyword_count=row[2]
        keyword=row[1]
        
        if keyword_id not in keyword_map :
            keyword_map[keyword_id] = keyword
            
def getName(keyword_id):
    return keyword_map[keyword_id]
```

```python
keyword_tree = {}
root = '74' #"death toll"
root_count = 0
node_id = 0

keyword_tree[node_id] = {}
keyword_tree[node_id]['name'] = getName(root)
keyword_tree[node_id]['count'] = root_conf
keyword_tree[node_id]['children_list'] = []
keyword_tree[node_id]['parent'] = 'null'
keyword_tree[node_id]['children'] = []

def checkChildren(child_list):
    keyword_list = []
    for key in child_list:    
        keyword_list.append(key)
    
    return keyword_list
        

for seq in news_seq_nested:
    pos = 0
    if seq[0] == root :
        keyword_tree[0]['count'] += 1     
        pos += 1
        if (pos < len(seq)):
            new_list = [root]
            new_list.append(seq[pos])
            node_id = news_id_list[news_seq_nested.index(new_list)]      
            #print seq[pos], checkChildren(keyword_tree[0]['children'])
            if node_id not in checkChildren(keyword_tree[0]['children_list']):  
                keyword_tree[0]['children_list'].append(node_id)
                keyword_tree[0][node_id] = {}
                keyword_tree[0][node_id]['name'] = getName(seq[pos])
                keyword_tree[0][node_id]['count'] = 1
                keyword_tree[0][node_id]['parent'] = keyword_tree[0]['name']
                keyword_tree[0]['children'].append(keyword_tree[0][node_id])
            else :
                #print keyword_tree[0][node_id]
                keyword_tree[0][node_id]['count'] += 1
                pos +=1
                #check the third
                if (pos < len(seq)):
                    new_list.append(seq[pos])
                    keyword_tree[0][node_id]['children_list'] = []
                    keyword_tree[0][node_id]['children'] = []
                    node_id_2 = news_id_list[news_seq_nested.index(new_list)]
                    if seq[pos] not in checkChildren(keyword_tree[0]['children_list']):       
                        keyword_tree[0][node_id]['children_list'].append(node_id_2)
                        keyword_tree[0][node_id][node_id_2] = {}
                        keyword_tree[0][node_id][node_id_2]['name'] = getName(seq[pos])
                        keyword_tree[0][node_id][node_id_2]['count'] = 1
                        keyword_tree[0][node_id][node_id_2]['parent'] = keyword_tree[0][node_id]['name']
                        
                        
                        #keyword_tree[0]['children'].append(keyword_tree[0][node_id][node_id_2])
                        for idx in keyword_tree[0]['children']:
                            for key in idx :
                                if key == 'children':
                                    idx['children'].append(keyword_tree[0][node_id][node_id_2])
                                    
                    else :
                        #check the third
                        keyword_tree[0][node_id][node_id_2]['count'] +=1
                
       
```
