```python
from pyspark.mllib.fpm import FPGrowth,FPGrowthModel
import pandas as pd

df = pd.read_csv('bigfile.csv')
df.pop('Count')
```

```python
mydict = {}
for x in range(len(df)):
    currentid = df.iloc[x,0]
    currentvalue = df.iloc[x,1]
    mydict.setdefault(currentid, [])
    mydict[currentid].append(currentvalue)
```

```python
data = []
```

```python
for i, v in mydict.items():
    data.append(v)
```

```python
rdd = sc.parallelize(data,2)
model = FPGrowth.train(rdd, 0.4, 2)
sorted(model.freqItemsets().collect())
```
